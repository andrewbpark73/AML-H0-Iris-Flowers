# 03) Dataset Visualization

```python
fig, ax = plt.subplots(5,5, figsize=(20,20))

variables = ['sepal length in cm', 'sepal width in cm', 'petal length in cm', 'petal width in cm', 'class']
labels = variables

class_colors = {
    'Iris-setosa': 'red',
    'Iris-versicolor': 'green',
    'Iris-virginica': 'blue'
}

for i, var_i in enumerate(variables):
    for j, var_j in enumerate(variables):
        if i == j:
            ax[i, j].text(0.5, 0.5, labels[i], ha='center', va='center', transform=ax[i, j].transAxes)
        else:
            ax[i, j].scatter(df[var_j], df[var_i], s=20, color=[class_colors[c] for c in df['class']])
            

plt.show()
```
